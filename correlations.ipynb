```python
import json
from pathlib import Path

import pandas as pd
import plotly.express as px

ROOT = Path('..').resolve()
PROCESSED = ROOT / 'data' / 'processed'
OUTPUTS = ROOT / 'outputs'
```

```python
# Load and merge sample datasets
volunteer = pd.read_csv(PROCESSED / 'sample_volunteer_engagement.csv')
demo = pd.read_csv(PROCESSED / 'sample_demographics.csv')
funding = pd.read_csv(PROCESSED / 'sample_nonprofit_funding.csv')

merged = volunteer.merge(demo, on='zip_code').merge(
    funding[['organization_id', 'funding', 'revenue']], on='organization_id'
)
merged.head()
```

```python
# Correlation heatmap
numeric = merged.select_dtypes(include='number')
corr = numeric.corr()
fig = px.imshow(corr, text_auto='.2f', color_continuous_scale='RdBu_r', zmin=-1, zmax=1)
fig.show()
```

```python
# Key relationship: education vs retention
fig = px.scatter(merged, x='education_pct', y='retention_rate', trendline='ols',
                 title='Education vs Volunteer Retention')
fig.show()
```

```python
# Load pipeline correlation results
with open(OUTPUTS / 'correlations.json') as f:
    results = json.load(f)

pd.DataFrame(results['top_correlations'])
```
